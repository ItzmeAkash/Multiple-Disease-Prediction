# **Parkinson's Disease Predictaion**

Importing libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn import svm
import warnings
warnings.filterwarnings('ignore')
```

##Data collection

```python
#loading the dataset

data='/content/drive/MyDrive/Self Projects/Ml/Multiple Disease Prediction/Parkinson Disease Prediction/parkinsons.csv'
```

```python
park_df=pd.read_csv(data)
```

```python
#First 5 rows in dataset

park_df.head()
```

```python
#rows and columns
park_df.shape
```

```python
# cheching all info
park_df.info()
```

```python
#checking the null values
park_df.isnull().sum()
```

```python
#Getting some statistical data

park_df.describe()
```

```python
# distribution of the values
park_df.status.value_counts()
```

```python
# grouping the data based on the target variable
park_df.groupby('status').mean()
```

##Data preprocessing

```python
#Covert negative values into postive
park_df['spread1']=abs(park_df['spread1'])
```

```python
park_df['spread1']
```

```python
park_df.tail(1)
```

```python
#We didn't need the name columns so drop the column

park_df.drop(columns=['name'],inplace=True)
```

```python
park_df
```

```python
X=park_df.drop(columns=['status'])
y=park_df['status']
```

```python
X.head(2)
```

```python
y
```

##Splitting the data to training and testing

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2, stratify=y, random_state=2)
```

```python
X_train.shape,X_test.shape
```

**Data Standardzation**

```python
scaler=StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

##Model training

```python
model=svm.SVC(kernel='linear')
```

```python
#training the svm model

model.fit(X_train,y_train)
```

Model Evaluate

```python
#Accuracy score on training data

X_train_pred=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_pred,y_train)

print("Accuracy with training: ", training_data_accuracy)
```

```python
#Accuracy score on test data

y_pred=model.predict(X_test)
testing_data_accuracy=accuracy_score(y_pred,y_test)

print("Accuracy with testing : ", testing_data_accuracy)
```
